```python
#question no.1= main:mean

list=[12,16,13,17,15,19,18,11]
sum_list=sum(list)
count_list=len(list)
mean=sum_list/count_list
print(f"sum of list is:",sum_list)
print(f"length of list is:",count_list)
print(f"mean of list is:",mean)

```
