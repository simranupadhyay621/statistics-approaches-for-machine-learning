```python
# given is the dataset as a list of integers where some integers are repeated
# for each iteration in for loop print the no. of  times aecyh integer is seen and store this information in k:v format
data=[2,5,5,5,6,6,7]
freq={}  #empty dictionary
for x in data :
    freq[x]=freq.get(x,0)+1
    print(freq)
```
